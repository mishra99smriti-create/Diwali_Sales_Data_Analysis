# Exploratory Data Analysis

## Gender

```python
# Plotting a bar chart for Gender and it's count
```

```python
ax = sns.countplot(x = 'Gender',data = df,  hue='Gender', palette=['pink', 'blue'], legend=False)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()    

```

## hue use for specify coloumn and legends remove extra legend box

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
# From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

AGE

```python
ax = sns.countplot(data = df, x = 'Age Group', hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(
    x='Age Group',
    y='Amount',
    data=sales_age,
    hue='Age Group',
    palette='viridis',
    legend=False
)

plt.show()
```

```python
# From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

# State

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,4)})

sns.barplot(
    data=sales_state,
    x='State',
    y='Orders',
    hue='State',
    palette='Set2',
    legend=False
)

plt.show()
```

```python
# From the above graph we can findout that highest order get from Uttar Pradesh
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,5)})

sns.barplot(
    data=sales_state,
    x='State',
    y='Amount',
    hue='State',
    palette='tab10',
    legend=False
)

plt.show()
```

```python
# The chart shows the top 10 states with the highest sales amount. The state with the tallest bar has the top-most sales among all states.
```

# Marital Status


```python
sns.set(rc={'figure.figsize':(7,5)})

ax = sns.countplot(
    data=df,
    x='Marital_Status',
    hue='Marital_Status',
    palette='Set2',
    legend=False
)

for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
# The chart shows the number of married and unmarried customers.
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(
    data=sales_state,
    x='Marital_Status',
    y='Amount',
    hue='Gender',
    palette='Set2'
)

plt.show()
```

```python
# From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

# Occupation

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(
    data=sales_state,
    x='Occupation',
    y='Amount',
    hue='Occupation',
    palette='tab20',
    legend=False
)

plt.show()
```

```python
# The chart shows the total sales amount for different occupations. Customers from some occupations contribute more to sales compared to others,
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(
    data=df,
    x='Occupation',
    hue='Occupation',
    palette='tab20',
    legend=False
)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# The countplot displays the number of customers in each occupation category using different colors for better visualization and comparison.
```

# Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})

ax = sns.countplot(
    data=df,
    x='Product_Category',
    hue='Product_Category',
    palette='tab20',
    legend=False
)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(
    data=sales_state,
    x='Product_ID',
    y='Orders',
    hue='Product_ID',
    palette='tab10',
    legend=False
)

plt.show()
```

```python
# The chart displays the top 10 products with the highest number of orders.
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))

df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(
    kind='bar',
    color=['red', 'blue', 'green', 'orange', 'purple', 'pink', 'brown', 'cyan', 'yellow', 'gray']
)

plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))

df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(
    kind='bar',
    color=['red', 'blue', 'green', 'orange', 'purple', 'pink', 'brown', 'cyan', 'yellow', 'gray']
)

plt.show()
```

## Conclusion

The analysis shows that female customers contributed more to overall purchases and sales compared to male customers. Customers in the age group of 26–35 years were the most active buyers during Diwali sales. States such as Uttar Pradesh, Maharashtra, and Karnataka generated the highest sales and order volumes. Product categories related to food, clothing, and electronics were among the most purchased items, indicating high customer demand in these categories.
